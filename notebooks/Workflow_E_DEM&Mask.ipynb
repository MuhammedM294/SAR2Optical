```python
dem = imageio.imread(f"preprocess/dem/DEM.tif")
```

```python
dem.min(), dem.max()
```

```python
plt.figure(figsize=(15,10))
plt.imshow(dem, cmap='gray')
plt.show()
```

```python
dem = dem.astype(np.float32)
```

```python
mean = np.mean(dem)
std = np.std(dem)
print(mean, std)
```

```python
norm_dem = (dem - mean) / std
```

```python
norm_dem.min(), norm_dem.max()
```

```python
norm_dem.mean(), norm_dem.std()
```

```python
plt.figure(figsize=(15,10))
plt.imshow(norm_dem, cmap='gray')
plt.show()
```

```python
imageio.imwrite(f"processed/dem/DEM.tif", norm_dem)
```
